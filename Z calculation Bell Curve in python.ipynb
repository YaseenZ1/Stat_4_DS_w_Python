```python
# Importing libraries 
import numpy as np 
import matplotlib.pyplot as plt 

# A custom function to calculate 
# probability distribution function 
def pdf(x): 
	mean = np.mean(x) 
	std = np.std(x) 
	y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2)) 
	return y_out 
	
# To generate an array of x-values 
x = np.arange(-2, 2, 0.1) 

# To generate an array of 
# y-values using corresponding x-values 
y = pdf(x) 

# Plotting the bell-shaped curve 
#plt.style.use('seaborn') 
plt.figure(figsize = (6, 6)) 
plt.plot(x, y, color = 'black', 
		linestyle = 'dashed') 

plt.scatter( x, y, marker = 'o', s = 25, color = 'red') 
#plt.show()
x.mean()
```

```python
import numpy as np

def probability_distribution_function(x: np.ndarray) -> np.ndarray:
    """
    Calculate the probability distribution function (PDF) for a given array of values.

    Args:
        x (np.ndarray): An array of values for which to calculate the PDF.

    Returns:
        np.ndarray: An array representing the PDF values corresponding to the input x values.
    """
    mean = np.mean(x)
    std = np.std(x)
    y_out = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std ** 2))
    return y_out
```

```python
from scipy.stats import norm

X = (76.6, 73.8, 81.2, 88.0)
for x in X:
   mu = 90
   sig = 20
   # Calculate Z score
   Z = (x - mu)/sig
   # Calculate the p-value for a z-score
   p_value = norm.sf(abs(Z))
   # Rounding Z score and p-value and assigning it to new valriable to be used in print function
   Zscore = round(Z, 3)
   Pvalue = round(p_value, 4)
   print("X =",x,", Z score =",Zscore,", p-value =",Pvalue)
    
   
```
